__7.__ Yes (6 of them are). First, the number of genes was found earlier:  
  
`awk -F"\t" '$4 ~ /gene/' exercises.bed | wc -l`  
  
The number of unique names is:  
  
`awk -F"\t" '$4 ~ /gene/ {print $4}' exercises.bed | sort -u | wc -l`  
  
Occurrences of unique names can be counted like this: 
  
`awk -F"\t" '$4 ~ /gene/ {print $4}' exercises.bed  | sort | uniq -c ` 

Filtering for counts greater than 1 can be done using awk:
  
`awk -F"\t" '$4 ~ /gene/ {print $4}' exercises.bed  | sort | uniq -c | awk '$1>1'`
